```python
import pandas as pd
```

```python
df=pd.read_csv('CDSdata\V5 CDS Composites-31Oct18.csv', header=1)
filters = {'Ccy':['USD'], 'DocClause':['XR14'], 'CompositeCurveRating':['B','BB','BBB']}
df = df[df[list(filters)].isin(filters).all(axis=1)]
df = df[df['Tier'].isin(['SNRFOR'])]
print(df)
```

```python
tickerlist = df['Ticker'].tolist()
print(tickerlist)
```

```python
import os, glob
import os.path
```

```python
path = r'CDSdata'
files_list = glob.glob(path + '/*.csv') #check the folder

df1 = pd.DataFrame(columns=df.columns)
for i in files_list: #for loop for the files
    df = pd.read_csv(i, header=1)
    df = df[df['Ccy'].isin(['USD'])]
    df = df[df['DocClause'].isin(['XR14'])]
    #df= df[df['CompositeCurveRating'].isin(['B'])]
    df = df[df['Ticker'].isin(tickerlist)]
    df = df[df['Tier'].isin(['SNRFOR'])]
    df1 = pd.concat([df1,df], ignore_index=True)
```

```python
print(df1)
```

```python
dictionary = {}
for j in range(len(tickerlist)):
    tick = tickerlist[j]
    df = df1.copy()
    df = df[df['Ticker'].isin([tick])]
    dictionary[tickerlist[j]+'_dataframe']=df
```

```python
print(dictionary)
```

```python
datapoints = ['Spread6m', 'Spread1y','Spread2y', 'Spread3y','Spread4y','Spread5y','Spread7y','Spread10y','Spread15y','Spread20y','Spread30y']
sortingdictionary = {}
for i in range(len(dictionary)):
    sortingdictionary[tickerlist[i]+'_dataframe'] = dictionary[tickerlist[i]+'_dataframe'][datapoints].count(axis=1).sum()
```

```python
print(sortingdictionary)
```

```python
from collections import OrderedDict
from itertools import islice
sorted_dictionary = OrderedDict(sorted(sortingdictionary.items(), key=lambda x: x[1], reverse=True))
sorted_tickerlist = list(islice(sorted_dictionary, 15))
print(sorted_tickerlist)
```

```python
path = r'Data by Ticker'
for i in range(len(sorted_tickerlist)):
        fr = dictionary[sorted_tickerlist[i]].to_csv(os.path.join(path,sorted_tickerlist[i]+'.csv'))
```
